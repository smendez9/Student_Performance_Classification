### Metadata
Student Information  
 
Target Variable: `Grade Class`  
- `GradeClass`: Classification of students' grades based on GPA:  
0: 'A' (GPA >= 3.5)  
1: 'B' (3.0 <= GPA < 3.5)  
2: 'C' (2.5 <= GPA < 3.0)  
3: 'D' (2.0 <= GPA < 2.5)  
4: 'F' (GPA < 2.0)  
* `StudentID`: A unique identifier assigned to each student (1001 to 3392).   
* `Age`: The age of the students ranges from 15 to 18 years.  
* `Gender`: Gender of the students, where 0 represents Male and 1 represents Female.  
* `Ethnicity`: The ethnicity of the students, coded as follows:  
0: Caucasian  
1: African American  
2: Asian  
3: Other  
* `ParentalEducation`: The education level of the parents, coded as follows:  
0: None  
1: High School  
2: Some College  
3: Bachelor's  
4: Higher  
* `StudyTimeWeekly`: Weekly study time in hours, ranging from 0 to 20.  
* `Absences`: Number of absences during the school year, ranging from 0 to 30.  
* `Tutoring`: Tutoring status, where 0 indicates No and 1 indicates Yes.   
* `ParentalSupport`: The level of parental support, is coded as follows:  
0: None  
1: Low  
2: Moderate  
3: High  
4: Very High   
* `Extracurricular`: Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.  
* `Sports`: Participation in sports, where 0 indicates No and 1 indicates Yes.  
* `Music`: Participation in music activities, where 0 indicates No and 1 indicates Yes.  
* `Volunteering`: Participation in volunteering, where 0 indicates No and 1 indicates Yes.    
* `GPA`: Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, parental involvement, and extracurricular activities.nvolvement, and extracurricular activities.ties.

### Feature Engineering

```python
# lets begin by creating a dataframe for features and target

# select features
cols_to_drop = ['GPA','Music','Volunteering','Sports','Age','Gender','StudentID','GradeClass']
features = data.drop(cols_to_drop, axis = 1)

# select target
target = data['GradeClass']
```

```python
features.columns
```

```python
# split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = .2, random_state = 42)
```

```python
# lets create a pipeline to process the features 

# instantiate objects for OHE and StandardScaler

ohe = OneHotEncoder(drop='first', handle_unknown='ignore')
scaler = StandardScaler() 

# select numerical and categorial features
num_features = ['StudyTimeWeekly', 'Absences']
cat_features = ['Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular']

# create processor
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', ohe, cat_features),
        ('num', 'passthrough', num_features)  # 'passthrough' means no transformation
    ]
)
```

```python
# preprocess data
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)


X_train = pd.DataFrame(X_train_preprocessed, columns=preprocessor.get_feature_names_out())
X_test = pd.DataFrame(X_test_preprocessed, columns=preprocessor.get_feature_names_out())
```

```python
# view correlation between features
corr_features = X_train.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_features,
    annot=True,               # show the correlation coefficients
    cmap='viridis',          # color map for the heatmap
    center=0,                 # center the color map at zero
    linewidths=0.5,           # width of the lines that divide cells
    linecolor='gray',         # color of the lines that divide cells
    fmt='.2f',                # format for the annotation text
    cbar_kws={"shrink": .8}   # shrink the color bar
)

plt.title('Correlation Matrix: Processed Features')
plt.show()
```

There aren't any strong correlation between features that might imply multicollinearity so I can move on to model training.

### Model Selection

#### Cross Validation Scoring

```python
# view model performance

models = {
    'GBC': GradientBoostingClassifier(),
    'RF': RandomForestClassifier(),
    'ABC': AdaBoostClassifier()
}

score = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv = 5)
    score[name] = scores.mean()
    print(f"{name}: Mean CV Score: {scores.mean()}")
```

```python
f1 = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv = 5, scoring = 'f1_micro')
    f1[name] = scores.mean()
    print(f"{name}: Mean CV F1_Micro: {scores.mean()}")
```

```python
class_distribution = y_train.value_counts()
print(class_distribution)

# Optionally, plot the distribution
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x=y_train)
plt.title('Class Distribution')
plt.show()
```

There is clearly come class imbalance.

```python
# handle class imbalance using SMOTE to create samples for minority class
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
# plot the distribution
class_distribution = y_resampled.value_counts()
print(class_distribution)

import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x=y_resampled)
plt.title('Class Distribution')
plt.show()
```

```python
# check f1 scores for each model with balanced classes

f1 = {}
for name, model in models.items():
    scores = cross_val_score(model, X_resampled, y_resampled, cv = 5, scoring = 'f1_macro')
    f1[name] = scores.mean()
    print(f"{name}: Mean CV f1_macro: {scores.mean()}")
```

```python
# Check score for each model with balanced classes
score = {}
for name, model in models.items():
    scores = cross_val_score(model, X_resampled, y_resampled, cv = 5)
    score[name] = scores.mean()
    print(f"{name}: Mean CV Score: {scores.mean()}")
```

Based on mean score and mean f1_micro score, RandomForestClassifier will be the model of choice.

#### Hyperparameter Tuning

```python
# create parameter dictionary for GridSearchCV
param_grid = {
              'max_depth': [6,16,32],
              'n_estimators': [10,90,300],
              'min_samples_split':[4,6,8]
             }

# perform grid seach and store in variable 
grid_search_cv_rf = GridSearchCV(RandomForestClassifier(random_state=42),
                              param_grid, verbose=0, cv=3)

%time grid_search_cv_rf.fit(X_resampled, y_resampled) # view how long the search takes
print("The best parameters are: ", grid_search_cv_rf.best_params_)
```

```python
# second round of hyperparameter tuning
param_grid = {
              'max_depth': [25,32,40],
              'n_estimators': [300,500,700],
              'min_samples_split':[3,4,5]
             }

# perform grid seach and store in variable 
grid_search_cv_rf = GridSearchCV(RandomForestClassifier(random_state=42),
                              param_grid, verbose=0, cv=3)

%time grid_search_cv_rf.fit(X_resampled, y_resampled)
print("The best parameters are: ", grid_search_cv_rf.best_params_)
```

```python
# view accuracy of the RandomForestClassifier
best_rfc = grid_search_cv_rf.best_estimator_

rfc_accuracy = best_rfc.score(X_test, y_test)

print("Random Forest Accuracy:", rfc_accuracy)
```

```python
# create function to plot confusion matrix
def plot_confusion_matrix(y,y_predict):

    cm = confusion_matrix(y, y_predict)

    # normalize the confusion matrix to get ratios
    cm_normalized = cm.astype(float) / cm.sum(axis=1)[:, np.newaxis]
    
    ax= plt.subplot()
    sns.heatmap(cm_normalized,
                fmt='d',
                annot=False, 
                ax = ax,
                cmap='viridis'); 
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix');  
    plt.show() 
```

```python
# create predictions and plot confusion matrix
preds = grid_search_cv_rf.predict(X_test)

plot_confusion_matrix(y_test,preds)
```

```python
accuracy = accuracy_score(y_test, preds)
precision = precision_score(y_test, preds, average = 'macro', zero_division=1)
recall = recall_score(y_test, preds, average = 'macro', zero_division=1)
f1 = f1_score(y_test, preds, average = 'macro', zero_division=1)
y_proba = best_rfc.predict_proba(X_test)
roc_auc = roc_auc_score(y_test, y_proba, multi_class='ovo')

print(f'Random Forest Classifier: \n Accuracy: {accuracy} \n Precision: {precision} \n Recall: {recall} \n F1: {f1} \n ROC_AUC: {roc_auc}')
```

```python
importances = best_rfc.feature_importances_

feature_names = preprocessor.get_feature_names_out()

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.title('Feature Importances from RandomForestClassifier')
plt.gca().invert_yaxis()
plt.show()
```
